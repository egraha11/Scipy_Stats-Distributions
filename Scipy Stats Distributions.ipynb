```python
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def graph(dist):
    fig, ax = plt.subplots(1,1)
    data = dist.rvs(size=1000)
    x = np.linspace(dist.ppf(0.01),dist.ppf(0.99), 100)
    
    
    sns.histplot(data=data, bins='auto')
    
    ax.set_xlim([x[0], x[-1]])

    #ax.legend(loc='best')

    plt.show()
    
    mean,var,skew,kurt = dist.stats(moments='mvsk')
    print(f"mean:{mean}")
    print(f"variance:{var}")
    print(f"skew:{skew}")
    print(f"kurtosis:{kurt}")
```

```python
#continuous dsitributions (loc and scale family)


#exponential distribution
#parameters - loc,scale

rv = stats.expon(scale=3)
graph(rv)
```

```python
#normal distribution
#parameters - loc,scale

rv = stats.norm(loc=3)
graph(rv)
```

```python
#lognormal distribution
#parameters - loc,scale, shape=st dev
rv = stats.lognorm(s=.8)
graph(rv)
```

```python
#uniform distribution
#parameters - loc,scale

rv = stats.uniform(scale=3, loc=5)
graph(rv)
```

```python
#weibull distribution
#parameters - loc,scale, shape

rv = stats.weibull_min(c=2, scale=3)
graph(rv)
```

```python
#gamma distribution
#parameters - loc,scale, shape=number of events

rv = stats.gamma(a=3, scale=1)
graph(rv)
```

```python
#beta distribution
#parameters - loc,scale, shape a, shape b

rv = stats.beta(a=2, b=8)
graph(rv)

rv = stats.beta(a=20, b=80)
graph(rv)
```

```python
#discrete distributions
#Poisson distribution
#parameters- shape

rv = stats.poisson(mu=3)

graph(rv)
```

```python
#Binomial 
#parameters shape p-probability n-number independent trials

rv = stats.binom(p=.4, n=5)

graph(rv)
```

```python
#hypergeometric
#parameters M - population size n-number of type 1 objects N-number of samples drawn

rv = stats.hypergeom(M=20, n=7, N=12)

graph(rv)
```

```python
#geometric
#parameters p - probability of success

rv = stats.geom(p=.4)

graph(rv)
```

```python
#Bernoulli
#parameters p - probability of success

rv = stats.bernoulli(p=.4)

graph(rv)
```

```python
#fit a distribution 
#from scipy stats - We believe that the parameter n was fewer than 30, 
#and we know that the parameter p must lie on the interval [0, 1]. 
#We record this information in a variable bounds and pass this information to fit.

rv = stats.weibull_min(c=2, scale=3)

data = rv.rvs(size=1000)

dist = stats.weibull_min

bounds = {'c': (0, 5), 'scale':(0, 5)}

res = stats.fit(dist, data, bounds)

print(res.params)

res.plot()
```
